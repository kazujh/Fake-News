```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
def sigmoid(n):
    return 1 / (1 + np.exp(-n))
```

```python
def initialize_weights(n_features):
    weights = np.zeros(n_features)
    bias = 0
    return weights, bias
```

```python
def predict(X, weights, bias):
    linear_model = X.dot(weights) + bias
    predictions = sigmoid(linear_model)
    return predictions
```

```python
def calculate_loss(y_true, y_pred):
    n = len(y_true)
    loss = (-1/n) * np.sum(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))
    return loss
```

```python
def gradient_descent(X, y, weights, bias, lr):
    n = X.shape[0]

    y_pred = predict(X, weights, bias)

    dw = X.T.dot(y_pred - y) / n
    db = np.sum(y_pred - y) / n

    weights -= lr * dw
    bias -= lr * db

    return weights, bias
```

```python
def train(X, y, lr=.1, epochs=1000, batch_size=500):
    n_features = X.shape[1]

    weights, bias = initialize_weights(n_features)

    losses = []

    for epoch in range(epochs):
        for i in range(0, X.shape[0], batch_size):
            X_batch = X[i:i + batch_size]
            y_batch = y[i:i + batch_size]
            
            
            weights, bias = gradient_descent(X_batch, y_batch, weights, bias, lr)

        y_pred = predict(X, weights, bias)
        loss = calculate_loss(y, y_pred)
        losses.append(loss)

        if epoch % 100 == 0:
            print(f"Epoch {epoch}, Loss: {loss}")
    return weights, bias, losses
```

```python
def classify(X, weights, bias, threshold=.5):
    probabilities = predict(X, weights, bias)
    return [1 if p >= threshold else 0 for p in probabilities]
```

```python
def confusion_matrix(y_true, y_pred):
    TP = FN = FP = TN = 0
    for true, pred in zip(y_true, y_pred):
        if true == 1:
            if pred == 1:
                TP += 1
            else:
                FN += 1
        else:
            if pred == 0:
                TN += 1
            else:
                FP += 1
    cm = np.array([[TP, FN], [FP, TN]])
    return cm
```

```python
def split_dataset(X, y, feature, threshold):
    ind1 = X[:, feature] <= threshold
    ind2 = X[:, feature] > threshold
    return X[ind1], y[ind1], X[ind2], y[ind2]
```

```python
def gini_impurity(groups, classes):
    n_instances = float(sum([len(group) for group in groups]))
    gini = 0.0
    for group in groups:
        n_samples = float(len(group))
        if n_samples == 0:
            continue
        score = 0.0
        for class_val in classes:
            proportion = [row['truth'] for row in group].count(class_val) / n_samples
            score += proportion ** 2
        gini += (1.0 - score) * (n_samples / n_instances)
    return gini
```

```python
def create_random_groups(dataset):
    shuffled_data = dataset.sample(frac=1, random_state=42).reset_index(drop=True)

    split_index = len(dataset) // 2

    group1 = shuffled_data.iloc[:split_index].todict('group1')
    group2 = shuffled_data.iloc[split_index:].todict('group2')

    return group1, group2
```
